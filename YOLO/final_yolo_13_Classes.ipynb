```python
# Step 6: Create the data.yaml file
data_yaml = """
path: /content/dataset
train: images/train
val: images/val

nc: 14
names:
  - Hello
  - Yes
  - I Love You
  - No
  - How
  - You
  - Thank You
  - Eat
  - Sleep
  - Water
  - Beat You
  - Iam
  - Want
  - Ok
"""

with open('/content/data.yaml', 'w') as f:
    f.write(data_yaml)
```

```python
# Step 7: Load the YOLOv8 model
model = YOLO('yolov8n.pt')  # Choose a pre-trained model as a starting point
```

```python
# Step 8: Train the model
model.train(data='/content/dataset/data.yaml', epochs=3, imgsz=224)
```

```python
# Step 9: Evaluate the model
metrics = model.val()
```
